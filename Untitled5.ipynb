```python
# parâmetros do modelo
beta = 0.96  # taxa de desconto intertemporal
sigma = 2.0  # coeficiente de aversão ao risco
alpha = 0.33  # participação do capital na produção
A = 1.0  # nível de tecnologia inicial
n = 0.01  # taxa de crescimento da população
delta = 0.05  # taxa de depreciação do capital
```

```python
# função de utilidade CRRA
def u(c):
    return c**(1-sigma)/(1-sigma)

# função de produção Cobb-Douglas com progresso tecnológico Hicks-neutro
def f(K, A):
    return A*K**alpha
```

```python
def rck_model(t, y):
    K, C, lambda_ = y
    A_t = A*np.exp(n*t)  # nível de tecnologia no tempo t
    Mp = -u(C) + beta*(1+n)*lambda_*u(C*np.exp(-t))  # condição de otimalidade para C
    Mlambda = A_t*alpha*K**(alpha-1) - (1+n+delta)*lambda_  # condição de otimalidade para lambda
    Mkp = f(K, A_t) - delta*K - C - (1+n)*K  # condição de transversalidade
    return [Mkp, Mp, Mlambda]
```

```python
# condições iniciais
K0 = 0.1
C0 = 0.1
lambda0 = 0.1
y0 = [K0, C0, lambda0]

# intervalo de tempo para simulação
t0 = 0.0
tf = 100.0
t_eval = np.linspace(t0, tf, 10000)
```

```python
sol = solve_ivp(rck_model, [t0, tf], y0, t_eval=t_eval, rtol=1e-9, atol=1e-9)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Parâmetros do modelo
alpha = 0.33    # participação do capital na produção
n = 0.01        # taxa de crescimento populacional
delta = 0.05    # taxa de depreciação do capital
sigma = 2       # aversão relativa ao risco

# Condições iniciais
K0 = 1
C0 = 0.8
lambda0 = 0

# Parâmetros da simulação
T = 100          # número de períodos
s_min = 0.01     # taxa de poupança mínima
s_max = 0.9      # taxa de poupança máxima
penalty = 100    # valor da penalidade para taxas de poupança fora do intervalo [s_min, s_max]

# Funções de utilidade e produção
def u(C):
    return (C**(1-sigma) - 1) / (1 - sigma)

def f(K, A):
    return A * K**alpha

# Equação de Euler-Lagrange
def euler_lagrange(K, C, lambda_, A, beta):
    s = 1 - C / f(K, A)
    Mlambda = A*alpha*K**(alpha-1) - (1+n+delta)*lambda_   # condição de otimalidade para lambda
    MC = u(C) - lambda_*s                            # condição de otimalidade para C
    MK = beta*lambda_*(A*alpha*K**(alpha-1) + (1-delta)) - lambda_ # condição de otimalidade para K
    return np.linalg.norm(np.array([Mlambda, MC, MK]))

# Simulação
K = np.zeros(T)
C = np.zeros(T)
lambda_ = np.zeros(T)
A = np.ones(T)

K[0] = K0
C[0] = C0
lambda_[0] = lambda0
beta = np.exp(-delta*T) # valor inicial para beta

for t in range(1, T):
    # Calcula a taxa de poupança ótima
    A[t] = A[t-1] * np.exp(0.02*np.random.randn())   # progresso tecnológico estocástico
    result = lambda x: euler_lagrange(x[0], x[1], x[2], A[t], beta)
    cons = ({'type': 'ineq', 'fun': lambda x: x[0]},  # K >= 0
            {'type': 'ineq', 'fun': lambda x: x[1]},  # C >= 0
            {'type': 'ineq', 'fun': lambda x: s_min - x[0]},  # K - sY >= 0
            {'type': 'ineq', 'fun': lambda x: x[0] - s_max})  # K - sY <= 0
    bnds = ((0, None), (0, None), (None, None))
    res = minimize(result, (K[t-1], C[t-1], lambda_[t-1]), method='SLSQP', bounds=bnds, constraints=cons)
    K[t], C[t], lambda_[t] = res.x
    beta *= np.exp(-delta)  # atualiza o valor de beta

# Gráfico da simulação
t = np.arange
```
